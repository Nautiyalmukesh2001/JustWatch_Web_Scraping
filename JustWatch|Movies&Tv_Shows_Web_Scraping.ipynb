````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
````

## **Scrapping Movies Data**

````python
# creating a function for urls in Movies Data and TV Shows data to make request and read through BeautifulSoup.
def fetch_movie_urls(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return "Failed to retrieve the page, status code:", response.status_code
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup


# justwatch movies url
url = 'https://www.justwatch.com/in/movies?release_year_from=2000'
movies_links =fetch_movie_urls(url)
````

## **Fetching Movie URL's**

````python
# creating a empty list to store the link of movies
movies_links_list = []

# Running a for loop to extract the movies link
for link in movies_links.find_all('a',attrs={'class':'title-list-grid__item--link'}):

  # using try-except block to catch the error if occured
  try:
    movies_links_list.append('https://www.justwatch.com' + link['href'])

    # print successful if no error ocuured with the movie link
    print("Successful",('https://www.justwatch.com' + link['href']))
  except:
    # print Error if error occured with the movie link
    print("Error",('https://www.justwatch.com' + link['href']))

    # if error occur then continue to next loop
    continue

  #  creating a sleep of 2 seconds to prevent overwhelming a website with requests and to mimic more natural browsing behavior
  # time.sleep(2) # uncomment this if getting error while executing code
````

## **Scrapping Movie Title**

````python
# creating an empty list to store the movies title
movies_title = []

# running a for loop in movies_links_list
for t in movies_links_list:

  # sending i(link) to the function to fetch html data
  title = fetch_movie_urls(t)

  # using try-except to catch the error
  try:

    # fetch the movie title and append the movies_title list
    movies_title.append(title.find_all('h1')[0].text.split('(')[0])
    print("successful",t)

  except:
    print('error',t)

    # if error occur skip to next loop
    continue

  # creating the sleep
  # time.sleep(2) # uncomment this if getting error while executing code

````

## **Scrapping release Year**

````python
# creating a empty list to store release year of movies
release_year_list = []

# running a for loop
for r in movies_links_list:
  release_year = fetch_movie_urls(r)

  # try-except block to catch error
  try:
    release_year_list.append(eval(release_year.find_all('h1')[0].text.split('(')[-1][:-1]))
    print("successful",r)
  except:
    print('error',r)
    continue

  # sleep
  # time.sleep(2) # uncomment this if getting error while executing code
````

## **Scrapping Genres**

````python
# creating an empty list to store the genre of movies
genres_list = []

# running a for loop
for g in movies_links_list:

  # sending the link to the function to fetch html data
  genres = fetch_movie_urls(g)

  # try-except block to catch the error
  try:

    # creating a found var to make ensure if i don't get the genres details so i will put 'Null' to corresponding link.So initialize False that we dont get genres.
    found = False

    # running a for loop after fetching 'div' tag and 'detail infos' class html data to find genre
    for g2 in genres.find_all('div',attrs={'class':'detail-infos'}):

      # conditional statement to check if 'detail-infos__subheading' class has genres
      if g2.find_all('h3',attrs={'class':'detail-infos__subheading'})[0].text.strip() == 'Genres':
        genres_list.append(g2.find_all('div',attrs={'class':'detail-infos__value'})[0].text.strip())


        # it means we found the genres
        found = True

        print("successful",g)
        # creating a break statement to exit the loop when we get the genres and to prevent the duplicates as 'detail infos' class have a repeated information of genres
        break

    # checking if genres not found
    if not found:
      genres_list.append('Null')

  except:
    print('error',g)
    continue

  # sleep
  # time.sleep(2) # uncomment this if getting error while executing code
````

## **Scrapping IMBD Rating**

````python
# creating an empty list to store the rating of movies
imdb_rating_list = []

# running a for loop
for ir in movies_links_list:
  imdb_rating = fetch_movie_urls(ir)

  # try-except block to catch the error
  try:
    imdb_rating_list.append(eval(imdb_rating.find_all('div',attrs={'class':'jw-scoring-listing__rating'})[0].text.split(' ')[0]))
    print("successful",ir)


  except:
    print('error',ir)
    continue

  # sleep
  # time.sleep(2) # uncomment this if getting error while executing code
````

## **Fetching Production Countries Details**

````python
# creating an empty list to store the production country of movies
production_country_list = []

# running a for loop
for pc in movies_links_list:

  # sending the link to the function to fetch html data
  production_country = fetch_movie_urls(pc)

  # try-except block to catch the error
  try:
    # creating a found var to make ensure if i don't get the production county details so i will put 'Null' to corresponding link.So initialize False that we dont get production country.
    found = False

    # running a for loop after fetching 'div' tag and 'detail infos' class html data to find production country
    for pc2 in production_country.find_all('div',attrs={'class':'detail-infos'}):

      # conditional statement to check if 'detail-infos__subheading' class has production country
      if pc2.find_all('h3',attrs={'class':'detail-infos__subheading'})[0].text.strip() == 'Production country':
        production_country_list.append(pc2.find_all('div',attrs={'class':'detail-infos__value'})[0].text.strip())

        # it means we found the genres
        found = True

        print("successful",pc)
        # creating a break statement to exit the loop when we get the production country and to prevent the duplicates as 'detail infos' class have a repeated information of production country
        break

    # checking if genres not found
    if not found:
      production_country_list.append('Null')

  except:
    print('error',pc)
    continue

  # time.sleep(2) # uncomment this if getting error while executing code
````

## **Fetching Streaming Service Details**

````python
# creating an empty list to store streaming service provider name
streaming_service_list = []

# running a for loop
for sv in movies_links_list:

  # sending link to function to fetch the data
  streaming_service = fetch_movie_urls(sv)

  # creating a set as there are more than one streaming service provider and using set to handle duplicates
  str_serv = []

  # try-except block
  try:
    for sv2 in streaming_service.find_all('div',attrs={'class':'buybox'})[0].find_all('img'):
      str_serv.append(sv2['alt'])
    streaming_service_list.append(str_serv)
    print("successful",sv)


  except:
    streaming_service_list.append('Null')
    print('error',sv)
    continue

  # sleep
  # time.sleep(2)  # uncomment this if getting error while executing code
````
